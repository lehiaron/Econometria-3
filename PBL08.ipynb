```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pyreadr
import numpy as np

# Base de dados 
# Carregar os dados do arquivo RData
result = pyreadr.read_r("Base3.RData")
df = list(result.values())[0]

# Certificar que 'region' é categórica
df['region'] = df['region'].astype(str).astype('category')

# Verificar se há valores ausentes na coluna 'region' ou 'fte' e remover
df = df.dropna(subset=['region', 'fte', 'tratado', 'pos', 'did', 'hrsopen', 'nregs', 'chain'])

# Regressões

# Regressão com erros padrão robustos a clusters
reg1 = smf.ols("fte ~ tratado + pos + did + hrsopen + nregs + C(chain)", data=df).fit()

# Regressão com erros padrão robustos a clusters
reg = smf.ols("fte ~ tratado + pos + did + hrsopen + nregs + C(chain)", data=df).fit(
    cov_type='cluster', cov_kwds={'groups': df['region'].cat.codes}  # Usando os códigos das categorias
)

# Exibir os resultados
print(reg.summary())
print(reg1.summary())
```
