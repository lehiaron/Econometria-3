```python
# Realizar regressão para cada valor de "TIME"
for time, group in time_groups:
    # Filtrar valores ausentes ou infinitos de "UNION" e "EARNINGS"
    group_clean = group[~group[["UNION", "EARNINGS"]].isna().any(axis=1)]
    group_clean = group_clean[~group_clean[["UNION", "EARNINGS"]].isin([float("inf"), float("-inf")]).any(axis=1)]
    
    # Verificar se há registros suficientes para a regressão
    if len(group_clean) > 1:  # Precisamos de mais de 1 observação para ajustar o modelo
        # Preparar os dados para a regressão
        X = group_clean["UNION"]  # Variável explicativa
        y = group_clean["EARNINGS"]  # Variável dependente
        
        # Adicionar uma constante (intercepto) ao modelo
        X = sm.add_constant(X)
        
        # Ajustar o modelo de regressão
        model = sm.OLS(y, X).fit()
        
        # Exibir o resumo da regressão para cada valor de "TIME"
        print(f"Regressão para o tempo {time}:")
        print(model.summary())
        print("\n")
    else:
        print(f"Tempo {time} não possui registros suficientes para a regressão.\n")
```

```python
# Concatenar todos os grupos corretamente
data_pooled = pd.concat([group for _, group in time_groups])

# Filtrar valores ausentes ou infinitos de "UNION" e "EARNINGS"
data_pooled = data_pooled.dropna(subset=["UNION", "EARNINGS"])
data_pooled = data_pooled[~data_pooled[["UNION", "EARNINGS"]].isin([float("inf"), float("-inf")]).any(axis=1)]

# Agrupar por tempo e por pessoa (ID)
data_pooled = data_pooled.groupby(["ID", "TIME"]).mean().reset_index()

# Verificar se há registros suficientes para a regressão
if len(data_pooled) > 1:
    # Preparar os dados para a regressão
    X = data_pooled["UNION"]  # Variável explicativa
    y = data_pooled["EARNINGS"]  # Variável dependente
    
    # Adicionar uma constante (intercepto) ao modelo
    X = sm.add_constant(X)
    
    # Ajustar o modelo de regressão
    model_pooled = sm.OLS(y, X).fit()
    
    # Exibir o resumo da regressão empilhada
    print("Regressão Pooled OLS:")
    print(model_pooled.summary())
else:
    print("Não há registros suficientes para a regressão Pooled OLS.")
```

```python
# Concatenar todos os grupos corretamente
data_pooled = pd.concat([group for _, group in time_groups])

# Filtrar valores ausentes ou infinitos das variáveis relevantes
variables = ["ID", "UNION", "EARNINGS", "AGE", "MALE", "S", "HOURS", "EXP"]
data_pooled = data_pooled.dropna(subset=variables)
data_pooled = data_pooled[~data_pooled[variables].isin([float("inf"), float("-inf")]).any(axis=1)]

# Agrupar por tempo e por pessoa (ID)
data_pooled = data_pooled.groupby(["ID", "TIME"]).mean().reset_index()

# Verificar se há registros suficientes para a regressão
if len(data_pooled) > 1:
    # Preparar os dados para a regressão
    X = data_pooled[["UNION", "AGE", "MALE", "S", "HOURS", "EXP"]]  # Variáveis explicativas
    y = data_pooled["EARNINGS"]  # Variável dependente
    
    # Adicionar uma constante (intercepto) ao modelo
    X = sm.add_constant(X)
    
    # Ajustar o modelo de regressão
    model_pooled = sm.OLS(y, X).fit()
    
    # Exibir o resumo da regressão empilhada
    print("Regressão Pooled OLS com variáveis de controle:")
    print(model_pooled.summary())
    
    # Calcular a matriz de correlação
    correlation_matrix = data_pooled[[ "UNION", "EARNINGS", "AGE", "MALE", "S", "HOURS", "EXP"]].corr()
    print("Matriz de correlação:")
    print(correlation_matrix)
    
    # Plotar o heatmap da correlação
    plt.figure(figsize=(10, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
    plt.title("Heatmap da Matriz de Correlação")
    plt.show()
else:
    print("Não há registros suficientes para a regressão Pooled OLS.")
```
